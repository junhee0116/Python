```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 66)

train_accuracy = []
test_accuracy = []
# 1에서 10까지 n_neighbors 적용
neighbors_setting = range(1, 11)
for n_neighbors in neighbors_setting:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors).fit(X_train, y_train)
    # 훈련세트 정확도 저장
    train_accuracy.append(clf.score(X_train, y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_setting, train_accuracy, label = 'train_accuracy')
plt.plot(neighbors_setting, test_accuracy, label = 'test_accuracy')
plt.ylabel('Accuracy')
plt.xlabel('n_neighbors')
plt.legend()
```

##### 리지 회귀 (Rigde)
<p>가중치의 절댓값을 가능한 작게 만들어 w의 모든 원소가 0에 가깝게 만드는 것입니다. 직관적으로, 모든 특성이 출력에 주는 영향을 최소한으로 만듭니다 (기울기를 작게 만듭니다)</p>
<p> 이런 제약을 <b> 규제(Regularization)</b>라고 합니다 <p>
<p> 규제란 과대적합이 되지 않도록 모델을 강제로 제한한다는 의미이고 리지 회귀에 사용하는 방식은 L2 규제라고 한다 </p>

```python
from sklearn.linear_model import Ridge
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

ridge = Ridge().fit(X_train, y_train)

print('train set score: {:.2f}'.format(ridge.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge.score(X_test, y_test)))
```

```python
# alpha parameter는 모델을 얼마나 단순화할지 지정
# alpha value가 낮으면 계수에 대한 제약이 풀리고 높으면 계수에 대한 제약이 강해진다
ridge10 = Ridge(alpha = 10).fit(X_train, y_train)

print('train set score: {:.2f}'.format(ridge10.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha = .1).fit(X_train, y_train)

print('train set score: {:.2f}'.format(ridge01.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge01.score(X_test, y_test)))
```

```python
# coef_ 크기 비교
plt.plot(ridge10.coef_, '^', label = 'alpha = 10')
plt.plot(ridge.coef_, 's', label = 'alpha = 1')
plt.plot(ridge01.coef_, 'v', label = 'alpha = 0.1')
plt.plot(lr.coef_, 'o')

plt.xlabel('coef list')
plt.ylabel('coef size')
xlims = plt.xlim()
# 수평선 plt.hlines(y, xmin, xmax)
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-20, 20)
plt.legend()
plt.show()
```

##### 라소 회귀 (Lasso)
<p>라소를 사용할 때 어떤 계수는 모델에서 완전히 제외되어 0이 됩니다. 어떻게 보면 특성 선택(feature selection)이 자동으로 이뤄진다고 볼 수 있습니다</p>
<p>일부 계수를 0으로 만들면 모델을 이해하기 쉬워지고 모델의 가장 중요한 특성이 무엇인지 드러내줍니다<p>
<p> 라소 회귀에 사용하는 방식은 L1 규제라고 한다 </p>

```python
from sklearn.linear_model import Lasso
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

lasso = Lasso().fit(X_train, y_train)

print('train set score: {:.2f}'.format(lasso.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso.score(X_test, y_test)))
print('using feature:', np.sum(lasso.coef_ != 0))
```

```python
# alpha parameter는 얼마나 강하게 0으로 보낼지를 조절
# alpha value가 낮으면 계수에 대한 제약이 풀리고 높으면 계수에 대한 제약이 강해진다
lasso001 = Lasso(alpha = .01, max_iter = 50000).fit(X_train, y_train)

print('train set score: {:.2f}'.format(lasso001.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso001.score(X_test, y_test)))
print('using feature:', np.sum(lasso001.coef_ != 0))
```

```python
lasso00001 = Lasso(alpha = .0001, max_iter = 50000).fit(X_train, y_train)

print('train set score: {:.2f}'.format(lasso00001.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso00001.score(X_test, y_test)))
print('using feature: ', np.sum(lasso00001.coef_ != 0))
```

```python
# coef_ 크기 비교
plt.plot(lasso.coef_, '^', label = 'alpha = 1')
plt.plot(lasso001.coef_, 's', label = 'alpha = 0.01')
plt.plot(lasso00001.coef_, 'v', label = 'alpha = 0.0001')
plt.plot(ridge01.coef_, 'o')

plt.xlabel('coef list')
plt.ylabel('coef size')
xlims = plt.xlim()
# 수평선 plt.hlines(y, xmin, xmax)
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-10, 10)
plt.legend()
plt.show()
```

##### 분류용 선형 모델
<p> 분류용 선형 모델에서는 결정 경계가 입력의 선형 함수입니다.</p>
<p> 가장 널리 알려진 두 개의 선형 분류 알고리즘은 Logistic Regression, SVM (Support Vector Machine)</p>
<p> 두 모델 다 L2 규제를 사용합니다. 규제의 강도를 결정하는 매개변수는 C입니다.</p>
<p> C의 값이 높아지면 규제가 감소합니다. 다시 말해 높은 C값을 설정하면 훈련 세트에 가능한 최대로 맞추려 하고, 낮추면 계수가 0에 가까워지도록 만듭니다 </p>
<p> C의 값이 낮아지면 데이터 다수에 맞추려고 하는 반면 C의 값이 높아지면 하나 하나의 데이터를 정확히 분류하려고 노력할 것입니다</p>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize = (8, 3))

for model, ax in zip([LinearSVC(max_iter = 5000), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill = False, eps = .5, ax = ax, alpha = .7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax = ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
axes[0].legend()
plt.show()
```

```python
# 고차원에서는 분류에 대한 선형 모델이 매우 강력해지며 특성이 많아지면 과대적합되지 않도록 하는 것이 매우 중요
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 42)

logreg = LogisticRegression(max_iter = 5000).fit(X_train, y_train)

print('train set score: {:.3f}'.format(logreg.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg.score(X_test, y_test)))
```

```python
logreg100 = LogisticRegression(C = 100, max_iter = 10000).fit(X_train, y_train)

print('train set score: {:.3f}'.format(logreg100.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg100.score(X_test, y_test)))
```

```python
logreg001 = LogisticRegression(C = .01, max_iter = 10000).fit(X_train, y_train)

print('train set score: {:.3f}'.format(logreg001.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg001.score(X_test, y_test)))
```

```python
plt.plot(logreg100.coef_.T, '^', label = 'C = 100')
plt.plot(logreg.coef_.T, 'o', label = 'C = 1')
plt.plot(logreg001.coef_.T, 'v', label = 'C = 0.01')
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel('features')
plt.ylabel('coef size')
plt.legend()
plt.show()
```

```python
# L1 규제를 사용
for c, marker in zip([.001, 1, 100], ['o', 'v', '^']):
    lr_l1 = LogisticRegression(solver = 'liblinear', C = c, penalty = 'l1', max_iter = 10000).fit(X_train, y_train)
    print('C = {:.3f} 인 l1 로지스틱 회귀의 훈련 정확도: {:.2f}'.format(c, lr_l1.score(X_train, y_train)))
    print('C = {:.3f} 인 l1 로지스틱 회귀의 테스트 정확도: {:.2f}'.format(c, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label = "C : {:.3f}".format(c))

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel('features')
plt.ylabel('coef size')
plt.legend()
plt.show()
```

```python
# 다중 클래스 분류 일대다 방식
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state = 42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend(['class 0', 'class 1', 'class2'])
plt.show()
```

```python
# LinearSVC
linear_svm = LinearSVC().fit(X, y)
print('계수 배열의 크기: ', linear_svm.coef_.shape)
print('절편 배열의 크기: ', linear_svm.intercept_.shape)
```

```python
# 3개의 이진 분류기가 만든 경계 시각화
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c = color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend(['class0', 'class1', 'class2', 'class0 b', 'class1 b', 'class2 b'], loc = (1.01, 0.3))
plt.show()
```

```python
# 2차원 평면의 모든 포인트에 대한 예측 결과
mglearn.plots.plot_2d_classification(linear_svm, X, fill = True, alpha = .7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intecept, color in zip(linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c = color)
plt.legend(['class0', 'class1', 'class2', 'class0 b', 'class1 b', 'class2 b'], loc = (1.01, 0.3))
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.show()
```
